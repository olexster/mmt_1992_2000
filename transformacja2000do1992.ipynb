```
def godlo_2000_10k(u2000):
  godlo_1=math.floor(u2000[1]/1e6)
  godlo_2=math.floor((u2000[0]/1e3-4920)/5)
  godlo_3=math.floor(((u2000[1]-godlo_1*1e6)/1e3-332)/8)
  godlo=str(godlo_1)+'_'+str(godlo_2)+'_'+str(godlo_3)
  return(godlo)
  
```
